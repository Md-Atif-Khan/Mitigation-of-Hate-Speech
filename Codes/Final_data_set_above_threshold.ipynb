```python
import pandas as pd
import shutil
import os

# Paths
source_folder = 'D:\Codes\Python\HateSpeech_res\data_anti_racism' 
csv_file_with_refs = 'D:\Codes\Python\HateSpeech_res\Final dataset above threshold(Anti-Racism)\working_data.csv' 
destination_folder = 'D:\Codes\Python\HateSpeech_res\Final dataset above threshold(Anti-Racism)\Original dataset' 

df = pd.read_csv(csv_file_with_refs)
df['ref'].nunique()
```

```python
for file_name in df['ref']:
    # Get the base file name (e.g., thread_799_1342245390540804096.csv)
    base_file_name = os.path.basename(file_name)
    
    # Source file path
    source_file_path = os.path.join(source_folder, base_file_name)
    
    # Destination file path
    destination_file_path = os.path.join(destination_folder, base_file_name)
    
    # Check if the source file exists
    if os.path.exists(source_file_path):
        # Copy the file to the destination folder
        shutil.copy(source_file_path, destination_file_path)
        print(f"Copied: {base_file_name}")
    else:
        print(f"File not found: {base_file_name}")

print("All specified files have been copied.")
```
